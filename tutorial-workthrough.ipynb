```python
## import pandas as an alias for ease of use
import pandas as pd
```

```python
## different ways to read data using Pandas w/ different file formats
coffee = pd.read_csv('./warmup-data/coffee.csv')
olympics_data = pd.read_excel('./data/olympics-data.xlsx')
results = pd.read_parquet('./data/results.parquet')
bios = pd.read_csv('./data/bios.csv')
```

Accessing Data with Pandas

```python
##.loc() function
coffee.loc[:3, ["Day", "Units Sold"]]
```

```python
##iloc() works the same using only index vals
coffee.iloc[:4, [0, 1]]
```

```python
## can also use loc() to modify vals in the df
coffee.loc[1, "Units Sold"] = 10
coffee
```

```python
coffee.sort_values(["Units Sold", "Coffee Type"], ascending=[0, 1])
## sorts the data by Units sold(descending 0 = false) then By coffee type (asc, 1 = True)
```
